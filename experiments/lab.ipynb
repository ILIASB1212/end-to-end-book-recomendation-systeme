```python
ratings = pd.read_csv('ratings.csv', delimiter=';', encoding='latin-1', on_bad_lines='skip')
```

```python
ratings.head()
```

```python
ratings['User-ID'].value_counts()
```

```python
x=ratings['User-ID'].value_counts()>200
```

```python
x[x]
```

```python
y=x[x].index
y
```

```python
ratings=ratings[ratings["User-ID"].isin(y)]
```

```python
ratings.head()
```

```python
ratings_with_books=ratings.merge(book,on='ISBN')
```

```python
ratings_with_books
```

```python
number_of_rating=ratings_with_books.groupby("Book-Title")["Book-Rating"].count().reset_index()
number_of_rating
```

```python
final_rating=ratings_with_books.merge(number_of_rating,on="Book-Title")
```

```python
final_rating
```

```python
final_rating=final_rating[final_rating["Book-Rating_y"]>=50]
```

```python
final_rating.head()
```

```python
final_rating.drop_duplicates(['User-ID','Book-Title'],inplace=True)
```

```python
final_rating.head()
```

```python
pivote=pd.pivot_table(index="Book-Title",columns="User-ID",values="Book-Rating_x",data=final_rating).fillna(0)
```

```python
pivote
```

```python
from scipy.sparse import csr_matrix
csr_data=csr_matrix(pivote)
csr_data
```

```python
from sklearn.neighbors import NearestNeighbors
model=NearestNeighbors(metric='cosine',algorithm='brute')
```

```python
model.fit(csr_data)
```

```python
model.kneighbors(pivote.iloc[0,:].values.reshape(1,-1),n_neighbors=6)
```

```python
distance,index=model.kneighbors(pivote.iloc[4,:].values.reshape(1,-1),n_neighbors=6)
```
